#  Sine and Cosine Plot
- Task: Plot $ \sin(x) $ and $ \cos(x) $ on the same graph for $ x $ values ranging from 0 to $ 2\pi $. Use different line styles, markers, and colors to distinguish between the two functions. Add a legend.

```python
import numpy as np
import matplotlib.pyplot as plt

# Defining the values of x
x = np.linspace(0, 2 * np.pi, 500)

# Putting the x into sin(x) and cos(x)
sin_x = np.sin(x)
cos_x = np.cos(x)

# Plotting the figures
plt.plot(x, sin_x, label=r'$\sin(x)$', color='blue', linestyle='-', marker='o', markersize=4, markevery=50)
plt.plot(x, cos_x, label=r'$\cos(x)$', color='green', linestyle='--', marker='s', markersize=4, markevery=50)

# Adding labels and title 
plt.title('Sine and cosine of x')
plt.xlabel('X')
plt.ylabel('f(x)')
plt.legend()
# plt.grid(True, linestyle="--", alpha=0.7)  # Also we can use grid 

# Final showing the result
plt.show()
```
