# Scatter Plot
- Task: Create a scatter plot of 100 random points in a 2D space. The x and y values should be randomly generated from a uniform distribution between 0 and 10. Use different colors and markers for the points. Add a title, axis labels, and a grid.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generating random numbers for x and y
x = np.random.uniform(0, 10, 100)
y = np.random.uniform(0, 10, 100)

# Random colors and Random markers
colors = np.random.rand(100)
markers = np.random.choice(['o', 's', 'd', '^', 'v', '<', '>', 'p', '*', 'h', 'H'], size=100)

# Plotting scatter plot
plt.figure(figsize=(8, 6))
for i in range(len(x)):
    plt.scatter(x[i], y[i], color=plt.cm.viridis(colors[i]), marker=markers[i], label=f'Point {i}' if i < 10 else '')

# Adding title, axis labels, and a grid
plt.grid(True, linestyle="--", alpha=0.7) # I used this style of grid to seems better
plt.title('100 random points')
plt.xlabel('X')
plt.ylabel('Y')

# Showing the result
plt.show()
```
