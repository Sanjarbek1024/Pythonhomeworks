# Subplots
### Task: Create a 2x2 grid of subplots. In each subplot, plot:

- Top-left: $ f(x) = x^3 $
- Top-right: $ f(x) = \sin(x) $
- Bottom-left: $ f(x) = e^x $
- Bottom-right: $ f(x) = \log(x+1) $ (for $ x \geq 0 $)

### Customize each plot with titles, axis labels, and different colors.

```python
import numpy as np
import matplotlib.pyplot as plt

# Defining the values of x
x1 = np.linspace(-10, 10, 500)
x2 = np.linspace(-2 * np.pi, 2 * np.pi, 500)
x3 = np.linspace(-2, 2, 500)
x4 = np.linspace(0, 10, 500)  

# Calculating the function values
y1 = x1**3
y2 = np.sin(x2)
y3 = np.exp(x3)
y4 = np.log(x4 + 1)

# Create a 2x2 grid of subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Top-left subplot: x^3
axs[0, 0].plot(x1, y1, color='blue', label='$f(x) = x^3$')
axs[0, 0].set_title('$f(x) = x^3$', fontsize=14)
axs[0, 0].set_xlabel('x', fontsize=12)
axs[0, 0].set_ylabel('f(x)', fontsize=12)
axs[0, 0].legend()
axs[0, 0].grid(True)

# Top-right subplot: sin(x)
axs[0, 1].plot(x2, y2, color='red', label='$f(x) = \sin(x)$')
axs[0, 1].set_title('$f(x) = \sin(x)$', fontsize=14)
axs[0, 1].set_xlabel('x', fontsize=12)
axs[0, 1].set_ylabel('f(x)', fontsize=12)
axs[0, 1].legend()
axs[0, 1].grid(True)

# Bottom-left subplot: e^x
axs[1, 0].plot(x3, y3, color='green', label='$f(x) = e^x$')
axs[1, 0].set_title('$f(x) = e^x$', fontsize=14)
axs[1, 0].set_xlabel('x', fontsize=12)
axs[1, 0].set_ylabel('f(x)', fontsize=12)
axs[1, 0].legend()
axs[1, 0].grid(True)

# Bottom-right subplot: log(x+1)
axs[1, 1].plot(x4, y4, color='purple', label='$f(x) = \log(x+1)$')
axs[1, 1].set_title('$f(x) = \log(x+1)$', fontsize=14)
axs[1, 1].set_xlabel('x', fontsize=12)
axs[1, 1].set_ylabel('f(x)', fontsize=12)
axs[1, 1].legend()
axs[1, 1].grid(True)

# Adjust layout for better spacing
plt.tight_layout()

# Show the plot
plt.show()
```
