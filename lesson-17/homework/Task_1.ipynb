# Basic Plotting
- Task: Plot the function $ f(x) = x^2 - 4x + 4 $ for $ x $ values between -10 and 10. Customize the plot with appropriate labels for the axes and a title.

```python
import numpy as np
import matplotlib.pyplot as plt

# Defining the values of x, y
x = np.linspace(-10, 10, 10000)
y = x**2 - 4*x + 4

# Creating plot
plt.plot(x, y, label="$f(x) = x^2 - 4x + 4$", color="blue")

# Adding labels and title
plt.xlabel("x", fontsize=12)
plt.ylabel("f(x)", fontsize=12)
plt.title("Plot of $f(x) = x^2 - 4x + 4$", fontsize=14)

# Adding grid and legend
# plt.grid(True, linestyle="--", alpha=0.7)
plt.legend(fontsize=12)

# Final step: Showing the graphs
plt.show()


```
