# Grouping and Aggregating

1. Grouped Aggregations on Titanic

- Group passengers by Pclass and calculate the following:
  -  Average age. 
  -  Total fare.
  -  Count of passengers.
- Save the results to a new DataFrame.

```python
import pandas as pd

# Reading excel file
titanic_df = pd.read_excel('data/data/titanic.xlsx')

# Grouping passengers
grouped_df = titanic_df.groupby('Pclass').agg(
    avg_age=('Age', 'mean'),
    total_fare=('Fare', 'sum'),
    passenger_count=('PassengerId', 'count')
).reset_index()

# Displaying results
grouped_df
```

2. Multi-level Grouping on Movie Data

- Group the movies by color and director_name.
- Find:
  - Total num_critic_for_reviews for each group.
  - Average duration for each group.

```python
import pandas as pd
import csv

# Reading movie.csv file
movie_df = pd.read_csv('data/data/movie.csv')

# Grouping the moviess
grouped_df = movie_df.groupby(['color', 'director_name']).agg({
    'num_critic_for_reviews': ['sum'],
    'duration': ['mean']
})

# Displaying the result
display(grouped_df)
```

3. Nested Grouping on Flights

- Group flights by Year and Month and calculate:
  - Total number of flights.
  - Average arrival delay (ArrDelay).
  - Maximum departure delay (DepDelay).


```python
import pandas as pd

# File name
files = 'data/flights'

# Loading files and joining them
df = pd.concat(map(pd.read_parquet, files), ignore_index=True)

# grouping by year and month
result = df.groupby(["Year", "Month"]).agg(
    TotalFlights=("Year", "count"),  
    AvgArrDelay=("ArrDelay", "mean"), 
    MaxDepDelay=("DepDelay", "max")  
).reset_index()

# Printing the result
print(result)

# Importing to csv
result.to_csv("output_flight_stats.csv", index=False)
```
